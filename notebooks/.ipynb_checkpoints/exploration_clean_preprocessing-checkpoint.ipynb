 # #day 2

```python
df = pd.read_csv("C:/Users/tarun/bank_churn_project/data/processed/data_exploration.csv")
```

```python
# 1.remove irrelevant columns
df.drop(['RowNumber','CustomerId','Surname'],axis =1,inplace =True)
```

```python
# 2. handle missing values 
if df.isnull().sum().sum() > 0 :
    df.fillna(df.median(numeric_only=True),inplace =True)
```

```python
df.head()
```

```python
#4. managing outliers
numeric_cols = df.select_dtypes(include =['float64','int64']).columns
for col in numeric_cols :
 Q1= df[col].quantile(0.25)
Q3 = df[col].quantile(0.75)
IQR = Q3-Q1
lower= Q1-1.5*IQR
upper= Q3 -1.5*IQR
df = df[(df[col]>=lower) & (df[col]<=upper)]
```

```python
#5.scaling 
from sklearn.preprocessing import StandardScaler
scalar= StandardScaler()
df[numeric_cols] = scalar.fit_transform(df[numeric_cols])
```

```python
df.to_csv("C:/Users/tarun/bank_churn_project/data/processed/Churn_Modelling_clean.csv", index=False)
```

## FEATURE ENGINEERING

```python
# CREATING AGE GROUPS
age_bins=[18,30,45,60,100]
age_labels = ['18-29','30-44','45-59','60+']
df['AgeGroup']= pd.cut(df['Age'],bins=age_bins ,labels= age_labels,right =False)

#CREATING TENURE GROUPS 
tenure_bins = [0, 3, 6, 10, 15]
tenure_labels = ['0-2', '3-5', '6-9', '10+']
df['TenureGroup'] = pd.cut(df['Tenure'], bins=tenure_bins, labels=tenure_labels, right=False)

#CRAETING BALANCE GROUPS 
balance_bins = [-1, 0, 0.5, 1.0, 2.5, df['Balance'].max() + 1]
balance_labels = ['No Balance', 'Very Low', 'Low', 'Medium', 'High']
df['BalanceGroup'] = pd.cut(df['Balance'], bins=balance_bins, labels=balance_labels, include_lowest=True)
```

```python
# INTERACTION FEATURES : Balance*Credit Score
df['Balence_CreditScore'] = df['Balance'] * df['CreditScore']
```

```python
# encoding categorical features
categorical_cols =['Gender', 'Geography','AgeGroup','TenureGroup','BalanceGroup']
df_featured_encoded =pd.get_dummies(df,columns = categorical_cols,drop_first = True)
```

```python
df_featured_encoded.to_csv("C:/Users/tarun/bank_churn_project/data/processed/bank_churn_eda_featured_encoded.csv", index=False)
```
